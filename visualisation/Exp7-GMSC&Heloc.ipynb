## GMSC

```python
original = pd.read_csv('../gmsc-training.csv')
new = pd.read_csv('../perfect_gmsc_counterfactuals_removed.csv', header=None)
counterfactual_origins = pd.read_csv('../perfect_gmsc_instances_removed.csv')
headers = original.columns.values.tolist()
reindexed_origins = counterfactual_origins.reindex(columns=headers)
new.columns = headers[1:]
reindexed_origins.to_csv('../good_order_origins_gmsc.csv', index=False)
```

```python
new_delta = []
origin_values = reindexed_origins.values[:,1:]
predictors = reindexed_origins.values[:,0]
for index, row in enumerate(new.values):
        diff = [a_i - b_i for a_i, b_i in zip(origin_values[index], row)]
        predictor = 0 if predictors[index] == 1 else 1
        diff.insert(0, predictor) 
        new_delta.append(diff)
        
new_delta_df = pd.DataFrame(new_delta, columns=headers)
desc = new_delta_df.describe()
desc.to_csv('delta_facts.csv')
new_delta_df.to_csv('../new_gmsc_delta.csv', index=False)
desc
```
