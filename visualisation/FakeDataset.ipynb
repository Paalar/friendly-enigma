```python
import pandas as pd

columns = ["X_0", "X_1", "X_2", "X_3", "X_4", "X_5"]

fake = pd.read_csv("../data/fake/fake_data_times_100.csv")
fake_c = pd.read_csv("../data/fake/fake_counterfactuals.csv", header=None, names=columns)
fake_d_c = pd.read_csv("../data/fake/fake_delta_counterfactuals.csv", header=None, names=["Targets", *columns])
fake_amp = pd.read_csv("../data/fake/augmented_counterfactuals_5_6.csv")
fake_amp_d = pd.read_csv("../data/fake/augmented_delta_counterfactuals_5_6.csv")
```

```python
def get_bar_count(dataset):
    column_count = dataset.astype(bool).sum(axis=0)
    count_sum = column_count.to_numpy().sum()
    ax = column_count.plot(kind="bar")

    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy() 
        ax.annotate(f'{(height/count_sum):.0%}', (x + width/2, y + height*1.02), ha='center')
```
