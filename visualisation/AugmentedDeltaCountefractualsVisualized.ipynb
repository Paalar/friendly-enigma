```python
import pandas as pd

delta = pd.read_csv("../data/augmented_delta_counterfactuals.csv", header=None)
target = delta.iloc[:,0]
delta = delta.iloc[:,1:]
```

```python
pd.options.display.max_columns = None
delta.describe(include='all')
```

```python
delta.mean().plot(kind="bar")
```

```python
column_count = delta.astype(bool).sum(axis=0)
column_count.plot(kind="bar")
```

```python
count_sum = column_count.to_numpy().sum()
ax = column_count.plot(kind="bar")

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{(height/count_sum):.0%}', (x + width/2, y + height*1.02), ha='center')
```
